```python
# Function to check if 500 is in the range
def is_500_in_range(range_str,final_rate):
    # Splitting the string to get the numerical range
    start, end = map(int, range_str[1:-1].split('-'))
    return final_rate >= start and final_rate <= end

def count_binary_events(dataset1, dataset2,event):
    # Ensure the date columns are in datetime format
    dataset1['Date'] = pd.to_datetime(dataset1['Date'])
    dataset1['FED_Date'] = pd.to_datetime(dataset1['FED_Date'])
    dataset2['datatime'] = pd.to_datetime(dataset2['datatime'])
    dataset2 = dataset2[dataset2['event_name']==event]

    # Count the FOMC meetings between dates
    counts = []
    for index, row in dataset1.iterrows():
        start_date = row['Date']
        end_date = row['FED_Date']
        count = dataset2[(dataset2['datatime'] > start_date) & (dataset2['datatime'] <= end_date)].shape[0]
        counts.append(count)

    count = pd.DataFrame(counts)
    count.columns = [event]
    dataset1 = pd.concat([dataset1,count],axis=1)

    return dataset1
        

def processing(data_path,fed_rates,meeting_date,binary_events):

    march_2023 = pd.read_csv(data_path + f'/CME Data/Data from CME (Excels)/Excels/FedMeeting_{meeting_date}.csv')
    non_zero_columns = march_2023.columns[1:][(march_2023.iloc[:, 1:].sum(axis=0) != 0).values].tolist()

    # Filter the dataframe to include only columns with non-zero values
    non_zero_df = march_2023[['Date'] + non_zero_columns]

    # Display the first few rows of the new dataframe
    non_zero_df['Date'] = pd.to_datetime(non_zero_df['Date'])
    non_zero_df.fillna(0, inplace=True)

    # Find the column name of the maximum number per date
    max_columns_per_date = non_zero_df.set_index('Date').idxmax(axis=1)
    max_columns_per_date = pd.DataFrame(max_columns_per_date)
    max_columns_per_date.columns = ['most_chosen']

    next_date = str(pd.to_datetime(meeting_date)+pd.Timedelta(days=1)).split(' ')[0]
    final_rate = fed_rates[fed_rates['DATE']==next_date]['DFEDTARU'].values[0]
    final_rate*=100
    max_columns_per_date['final_rate'] = final_rate
    max_columns_per_date['FED_Date'] = next_date

    # Applying the function to create a new column
    max_columns_per_date['is_correct'] = max_columns_per_date['most_chosen'].apply(lambda x: is_500_in_range(x,final_rate))
    max_columns_per_date['is_correct'] = max_columns_per_date['is_correct'].astype(int)

    max_columns_per_date = max_columns_per_date.reset_index()
    max_columns_per_date['days_until_meeting'] = pd.to_datetime(next_date) - max_columns_per_date['Date']
    max_columns_per_date['days_until_meeting'] = max_columns_per_date['days_until_meeting'].apply(lambda x: x.days)

    # count binary events 
    for event in binary_events['event_name'].unique():   
        print(event)
        max_columns_per_date = count_binary_events(max_columns_per_date, binary_events,event)

    # add aditional columns fix later 
    max_columns_per_date['number_of_rate_decisions_in_between'] = max_columns_per_date['FOMC Meeting'] - 1
    max_columns_per_date['number_of_rate_decisions_in_between'] = max_columns_per_date['number_of_rate_decisions_in_between'].apply(lambda x: 0 if x<0 else x)


    # kurtosis and skewness
    non_zero_df['row_skewness'] = non_zero_df[non_zero_columns].apply(lambda row: row.skew(), axis=1)
    non_zero_df['row_kurtosis'] = non_zero_df[non_zero_columns].apply(lambda row: row.kurtosis(), axis=1)
    non_zero_df['kurtosis_change'] = non_zero_df['row_kurtosis'].diff()
    non_zero_df['accumulated_kurtosis_change'] = non_zero_df['kurtosis_change'].cumsum()
    non_zero_df['skewness_change'] = non_zero_df['row_skewness'].diff()
    non_zero_df['accumulated_skewness_change'] = non_zero_df['skewness_change'].cumsum()
    max_columns_per_date = pd.merge(max_columns_per_date,non_zero_df[['Date','row_skewness','row_kurtosis','kurtosis_change','accumulated_kurtosis_change','skewness_change','accumulated_skewness_change']],on='Date',how='left')

    return max_columns_per_date
```

```python
meeting_dates = ['20230322','20230503','20230614','20230920','20230726']
meeting_dates = ['20230503']
df_full = pd.DataFrame()

for meeting_date in meeting_dates: 
    df = processing(data_path,fed_rates,meeting_date,binary_events)
    df_full = pd.concat([df_full,df],axis=0)
```

```python
# do we see a change in variable when an event happened 3 days before 3 days after 

# Function to get dynamic surrounding rows
def get_surrounding_rows(df, column_name, before=3, after=3):
    indices = df.index[df[column_name] == 1].tolist()
    surrounding_indices = []

    for index in indices:
        start = max(0, index - before)
        end = min(len(df), index + after + 1)  # +1 because range end is exclusive
        surrounding_indices.extend(range(start, end))

    # Remove duplicates and sort
    surrounding_indices = sorted(set(surrounding_indices))

    return df.iloc[surrounding_indices],indices

# Function to calculate percentage change between pairs of rows
def calculate_percentage_change(df, column_name):
    # Slices for alternating rows
    first_rows = df[column_name].iloc[::2].reset_index(drop=True)
    second_rows = df[column_name].iloc[1::2].reset_index(drop=True)

   # Calculate percentage change with condition for zeros
    percentage_change = np.where(
        (first_rows == 0) & (second_rows == 0),
        0,  # If both rows are zero, return 0
        (second_rows - first_rows) / first_rows * 100
    )
    return percentage_change

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
df = df_full 

# Preprocessing the data
# Dropping 'Date' and 'most_chosen' columns
df = df.drop(['Date', 'most_chosen','FED_Date','final_rate'], axis=1)

# Fill NaN values with 0 (or another strategy can be chosen)
df.fillna(0, inplace=True)

# Splitting the data into features and target
X = df.drop('is_correct', axis=1)
y = df['is_correct']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creating and training the RandomForestClassifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)

# Making predictions on the test set
y_pred = random_forest_model.predict(X_test)
y_proba = random_forest_model.predict_proba(X_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```
